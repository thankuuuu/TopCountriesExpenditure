```python
#Clean Data 

import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#import the dataset from each excel's data sheet without excel header description
#Military Spending
AllMilitarySpend = pd.read_excel('Military Spending Dataset.xlsx', header = 5, sheet_name = "Current USD", index_col = 0) 
AllMSPercentGDP = pd.read_excel('Military Spending Dataset.xlsx', header = 5, sheet_name = "Share of GDP", index_col = 0) 
AllMSPerCapita = pd.read_excel('Military Spending Dataset.xlsx', header = 6, sheet_name = "Per capita", index_col = 0)
AllMSPercent = pd.read_excel('Military Spending Dataset.xlsx', header = 7, sheet_name = "Share of Govt. spending", index_col = 0)
#GDP
AllGDP = pd.read_csv('GDP.csv', header = 2, index_col = 0)
#Health Spending
AllHSGDP = pd.read_csv('Current health expenditure of GDP.csv', header = 2, index_col = 0)
AllHSPerCapita = pd.read_csv('Current health expenditure per capita.csv', header = 2, index_col = 0)
#Education Spending
AllESGDP = pd.read_excel('Government expenditure on education of GDP.xls', header = 3, sheet_name = "Data",index_col = 0)

#Save 10 countries contain just only data from 2013 to 2018 in new files
#Note! MSPercentGDP and MSPercent in the excel file is %, when open in python it /100 to be a percent. Therefore have to compensate it by *100 
MilitarySpend = AllMilitarySpend.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],[2010,2011,2012,2013,2014]]
MSPercentGDP = (AllMSPercentGDP.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],[2010,2011,2012,2013,2014]])*100
MSPerCapita = AllMSPerCapita.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],[2010,2011,2012,2013,2014]]  
MSPercent = (AllMSPercent.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],[2010,2011,2012,2013,2014]])*100
GDP = AllGDP.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran, Islamic Rep.','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],['2010','2011','2012','2013','2014']] 
GDP.rename(index={'Iran, Islamic Rep.':'Iran'}, inplace=True)
HSGDP = AllHSGDP.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran, Islamic Rep.','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],['2010','2011','2012','2013','2014']]  
HSGDP.rename(index={'Iran, Islamic Rep.':'Iran'}, inplace=True)
HSPerCapita = AllHSPerCapita.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran, Islamic Rep.','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],['2010','2011','2012','2013','2014']]  
HSPerCapita.rename(index={'Iran, Islamic Rep.':'Iran'}, inplace=True)
ESGDP = AllESGDP.loc[['Argentina','Australia','Austria','Brazil','France','Germany','Indonesia','Italy','Iran, Islamic Rep.','Mexico','Netherlands','New Zealand','Norway','South Africa','Sweden'],['2010','2011','2012','2013','2014']]  
ESGDP.rename(index={'Iran, Islamic Rep.':'Iran'}, inplace=True)


#Check whether any datasets have null or not
MilitarySpend.info()
MSPercentGDP.info()
MSPerCapita.info()
MSPercent.info()
GDP.info()
HSGDP.info()
HSPerCapita.info()
ESGDP.info()
```

```python
# Creating columns name Country same as index
GDP['Country'] = GDP.index
HSGDP['Country'] = HSGDP.index
HSPerCapita['Country'] = HSPerCapita.index
ESGDP['Country'] = ESGDP.index

#Setting Country to be the index as others sheet
GDP.set_index('Country', inplace=True)
HSGDP.set_index('Country', inplace=True)
HSPerCapita.set_index('Country', inplace=True)
ESGDP.set_index('Country', inplace=True)

#Changing columns name from obj to int
GDP.columns = GDP.columns.astype(int)
HSGDP.columns = HSGDP.columns.astype(int)
HSPerCapita.columns = HSPerCapita.columns.astype(int)
ESGDP.columns = ESGDP.columns.astype(int)
```

```python
ESGDP
```

```python
#Change GDP unit from $ to Millions$
GDP = GDP/1000000

#Health Spending
HealthSpend = HSGDP*GDP/100
#Education Spending
EducationSpend = ESGDP*GDP/100

#Finding Population
Population = MilitarySpend*1000000/MSPerCapita

#Education Spending Per Capita = Education Spending % GDP * GDP Per Capita
ESPerCapita = EducationSpend*1000000/Population

#GDP Per Capita
GDPPerCapita = GDP*1000000/Population
```

```python
#Military Spending Growing
#Fixed
MSFixedGrowing = MilitarySpend.iloc[:,-1]-MilitarySpend.iloc[:,0]
#Percent 
MSPercentGrowing = (MilitarySpend.iloc[:,-1]-MilitarySpend.iloc[:,0])/MilitarySpend.iloc[:,0]*100

#Health Spending Growing
#Fixed
HSFixedGrowing = HealthSpend.iloc[:,-1]-HealthSpend.iloc[:,0]
#Percent 
HSPercentGrowing = (HealthSpend.iloc[:,-1]-HealthSpend.iloc[:,0])/HealthSpend.iloc[:,0]*100

#Education Spending Growing
#Fixed
ESFixedGrowing = EducationSpend.iloc[:,-1]-EducationSpend.iloc[:,0]
#Percent 
ESPercentGrowing = (EducationSpend.iloc[:,-1]-EducationSpend.iloc[:,0])/EducationSpend.iloc[:,0]*100
```

```python
#Creating Dataframe
Growing = pd.DataFrame(columns = ['MSFixedGrowing', 'MSPercentGrowing','HSFixedGrowing','HSPercentGrowing','ESFixedGrowing','ESPercentGrowing'])
Growing['MSFixedGrowing'] = MSFixedGrowing
Growing['MSPercentGrowing'] = MSPercentGrowing
Growing['HSFixedGrowing'] = HSFixedGrowing
Growing['HSPercentGrowing'] = HSPercentGrowing
Growing['ESFixedGrowing'] = ESFixedGrowing
Growing['ESPercentGrowing'] = ESPercentGrowing
```

```python
MeanSpend = pd.DataFrame(index=MilitarySpend.index)
MeanSpend['MilitarySpend'] = MilitarySpend.mean(axis=1)
MeanSpend['MSPercentGDP'] = MSPercentGDP.mean(axis=1)
MeanSpend['MSPerCapita'] = MSPerCapita.mean(axis=1)
MeanSpend['HealthSpend'] = HealthSpend.mean(axis=1)
MeanSpend['HSGDP'] = HSGDP.mean(axis=1)
MeanSpend['HSPerCapita'] = HSPerCapita.mean(axis=1)
MeanSpend['EducationSpend'] = EducationSpend.mean(axis=1)
MeanSpend['ESGDP'] = ESGDP.mean(axis=1)
MeanSpend['ESPerCapita'] = ESPerCapita.mean(axis=1)
MeanSpend['GDP'] = GDP.mean(axis=1)
MeanSpend['GDPPerCapita'] = GDPPerCapita.mean(axis=1)


#Calculating the mean of each sheets
MilitarySpend['Mean'] = MilitarySpend.mean(axis=1)
MSPercentGDP['Mean'] = MSPercentGDP.mean(axis=1)
MSPerCapita['Mean'] = MSPerCapita.mean(axis=1)
MSPercent['Mean'] = MSPercent.mean(axis=1)
GDP['Mean'] = GDP.mean(axis=1)
HealthSpend['Mean'] = HealthSpend.mean(axis=1)
HSGDP['Mean'] = HSGDP.mean(axis=1)
HSPerCapita['Mean'] = HSPerCapita.mean(axis=1)
EducationSpend['Mean'] = EducationSpend.mean(axis=1)
ESGDP['Mean'] = ESGDP.mean(axis=1)
ESPerCapita['Mean'] = ESPerCapita.mean(axis=1)
```

```python
writer = pd.ExcelWriter('Top16CountryExpenditures.xlsx', engine='xlsxwriter')
MilitarySpend.to_excel(writer, sheet_name='MilitarySpend(M$)')
MSPercentGDP.to_excel(writer, sheet_name='MSPercentGDP')
MSPerCapita.to_excel(writer, sheet_name='MSPerCapita($)')
MSPercent.to_excel(writer, sheet_name='MSPercent')
GDP.to_excel(writer, sheet_name='GDP(M$)')
HealthSpend.to_excel(writer, sheet_name='HealthSpend(M$)')
HSGDP.to_excel(writer, sheet_name='HSGDP')
HSPerCapita.to_excel(writer, sheet_name='HSPerCapita($)')
EducationSpend.to_excel(writer, sheet_name='EducationSpend(M$)')
ESGDP.to_excel(writer, sheet_name='ESGDP')
ESPerCapita.to_excel(writer, sheet_name='ESPerCapita($)')
Population.to_excel(writer, sheet_name='Population')
Growing.to_excel(writer, sheet_name='Growing')
MeanSpend.to_excel(writer, sheet_name='MeanSpend')
writer.save()
writer.close()
```
